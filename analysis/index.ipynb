# Computing and visualizing PCs using rsvd and ggplot2: an illustration with the Iris data

## Quick Start

1. Install the required packages from CRAN.

   ```R
   install.packages(c("repr","ggplot2","rsvd"))
   ```

2. Walk through [the first notebook](analysis/Compute_PCs.html) to compute the PCs using the `rsvd` package.

3. Walk through [the second notebook](analysis/Plot_PCs.html) to generate evocative visualizations of the PCA results using `ggplot2`.
